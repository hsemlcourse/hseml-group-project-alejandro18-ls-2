# 02 Baseline — Logistic Regression without FE

Baseline должен быть простой моделью без feature engineering. Он нужен как нижняя планка для последующих экспериментов.

```python
from pathlib import Path
from sklearn.linear_model import LogisticRegression

from src.data import load_dataset
from src.metrics import evaluate_model
from src.modeling import make_pipeline
from src.preprocessing import train_val_test_split

DATA_PATH = Path('../data/raw/online_shoppers_intention.csv')
X, y = load_dataset(DATA_PATH)
X_train, X_val, X_test, y_train, y_val, y_test = train_val_test_split(X, y)

baseline = make_pipeline(
    LogisticRegression(max_iter=2000, random_state=42),
    use_feature_engineering=False,
    clip_outliers=False,
)
baseline.fit(X_train, y_train)
evaluate_model(baseline, X_val, y_val)
```

Вывод: baseline сравнивается с моделями CP2 по PR-AUC, ROC-AUC и F1. Основная метрика — PR-AUC из-за дисбаланса target.
